```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# Upload image
uploaded = files.upload()
image_path = list(uploaded.keys())[0]

# Load and preprocess
img = cv2.imread(image_path)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# Binary threshold
_, binary = cv2.threshold(blurred, 50, 255, cv2.THRESH_BINARY_INV)

# Define pixel-to-nanometer conversion (you can adjust this value)
nm_per_pixel = 100  # Example: each pixel = 100 nanometers

# Initialize data
contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
fiber_lengths_px = []
fiber_lengths_nm = []
fiber_angles_deg = []
fiber_surface_area_nm2 = 0

# Draw contours and collect data
output = img.copy()
for i, cnt in enumerate(contours):
    # Length in pixels and nm
    length_px = cv2.arcLength(cnt, True)
    fiber_lengths_px.append(length_px)
    length_nm = length_px * nm_per_pixel
    fiber_lengths_nm.append(length_nm)

    # Area in nm²
    area_px = cv2.contourArea(cnt)
    fiber_surface_area_nm2 += area_px * (nm_per_pixel ** 2)

    # Orientation angle
    rect = cv2.minAreaRect(cnt)
    angle = rect[-1]
    if angle < -45:
        angle = 90 + angle
    fiber_angles_deg.append(angle)

    # Draw + label
    M = cv2.moments(cnt)
    if M["m00"] != 0:
        cx = int(M["m10"] / M["m00"])
        cy = int(M["m01"] / M["m00"])
        cv2.drawContours(output, [cnt], -1, (0, 255, 0), 2)
        label_text = f"{i+1}"
        cv2.putText(output, label_text, (cx, cy), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 0, 0), 2)

# Porosity
total_pixels = binary.size
fiber_pixels = np.sum(binary == 255)
porosity = 1 - (fiber_pixels / total_pixels)

# Total image area in nm²
image_height, image_width = binary.shape
total_area_px = image_height * image_width
total_image_area_nm2 = total_area_px * (nm_per_pixel ** 2)

# Display results
fig, axs = plt.subplots(1, 3, figsize=(18, 6))
axs[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axs[0].set_title("🔬 Original Image")
axs[0].axis("off")

axs[1].imshow(binary, cmap="gray")
axs[1].set_title("🧼 Binary Image")
axs[1].axis("off")

axs[2].imshow(cv2.cvtColor(output, cv2.COLOR_BGR2RGB))
axs[2].set_title("🧵 Analyzed Image")
axs[2].axis("off")

plt.tight_layout()
plt.show()

# Print detailed fiber results
print("📊 Fiber Measurements:")
for i, (px, nm, angle) in enumerate(zip(fiber_lengths_px, fiber_lengths_nm, fiber_angles_deg)):
    print(f"Fiber {i+1}: Length = {px:.2f} px / {nm:.2f} nm, Angle = {angle:.2f}°")

# Summary
print(f"\n🟤 Porosity: {porosity * 100:.2f}%")
print(f"🧩 Total Fiber Surface Area: {fiber_surface_area_nm2:.2f} nm²")
print(f"🌐 Total Image Surface Area: {total_image_area_nm2:.2f} nm²")

# Orientation histogram
plt.hist(fiber_angles_deg, bins=20, color='teal', edgecolor='black')
plt.title("📐 Fiber Orientation Histogram")
plt.xlabel("Angle (degrees)")
plt.ylabel("Count")
plt.grid(True)
plt.show()
```
